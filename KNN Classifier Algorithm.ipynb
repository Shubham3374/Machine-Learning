```python
# Importing important libraries



from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier , KNeighborsRegressor
from sklearn.model_selection import cross_val_score , train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Extracting the Datasets

irisData = load_iris()
```

```python
# printing the keys

irisData.keys()
```

```python
# EDA (Exploratory Data Analysis)

print(irisData.DESCR)
```

```python
# Creating input and output features :


X = irisData.data
y = irisData.target

# Printing the shape of the data 

print("X matrix dimensionality" , X.shape)
print("y matrix dimensionality" , y.shape)

```

```python
# Performing train and test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

```python
# Searching that for optimal value of k

# List of access from k range 

k_range = range(1,30)

# List to store accuracy
k_scores = []

# Looping throgh values of k

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(knn, X, y, cv = 10, scoring = 'accuracy')
    k_scores.append(scores.mean())
    
print("Cross validity accuracy :\n " , k_scores)
```

```python
k_scores
```

```python
# Getting optimal values of k

plt.plot(k_range,k_scores)
plt.xlabel("Values of K")
plt.ylabel("Cross validated accuracy")
plt.title("Optimal value of K ")
plt.show()
```

```python
# k = 13 is the optimal value of k
```

```python
# Creating the optimal model

knn = KNeighborsClassifier(n_neighbors = 13)

# Fitting the model

knn.fit(X_train ,y_train)
```

```python
# Predictions from the model


y_pred = knn.predict(X_test)
```

```python
# Getting the metrics

from sklearn import metrics
print("Accuracy :", metrics.accuracy_score(y_test,y_pred))
print("Precision :", metrics.precision_score(y_test,y_pred , average = 'micro'))
print("Recall :", metrics.recall_score(y_test,y_pred , average = 'micro'))
print("f1 Score :", metrics.f1_score(y_test,y_pred , average ='micro'))
```
